```python
static_kernel = np.ones((5,5))/25
print(static_kernel)
```

```python
img  = cv2.imread(r"/content/254695997_218070303773876_394825860655432974_n.jpg")
cv2_imshow( img)
filters = cv2.filter2D(img, -1, static_kernel)
cv2_imshow(filters)
plt.imshow(filters)
cv2.waitKey(0)
```

```python
gaussian_blur = cv2.GaussianBlur(img, (11,11), 16)
cv2_imshow(gaussian_blur)
cv2.waitKey(0)
```

```python
avg = cv2.blur(img, (5, 5))
cv2_imshow(avg)
cv2.waitKey(0)
```

```python
median_blur = cv2.medianBlur(img, 11)
cv2_imshow(median_blur)
cv2.waitKey(0)
```

```python
bilateral_filter = cv2.bilateralFilter(img, 17, 125, 125)
cv2_imshow(bilateral_filter)
cv2.waitKey(0)
```

```python
# import the opencv library
import cv2
import matplotlib.pyplot as plt
vid = cv2.VideoCapture(0)
  
while(vid.isOpened()):
      
    # Capture the video frame
    # frame
    ret, frame = vid.read()
    if ret:
        print(ret)
        cv2.imshow('frame', frame)
        plt.imshow(frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
  
# After the loop release the cap object
vid.release()
# Destroy all the windows
cv2.destroyAllWindows()
```
