```python
img = cv2.imread(r"/content/drive/MyDrive/Resources/bmw.jfif")
gra = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

canny_edge_LPF = cv2.Canny(gra, 25, 25)
cv2_imshow( canny_edge_LPF)

canny_edge_HPF = cv2.Canny(gra, 255, 255)
cv2_imshow(canny_edge_HPF)

cv2.waitKey(0)
```

```python
erosion = cv2.erode(canny_edge_HPF, (5,5), iterations=1)
cv2_imshow( canny_edge_HPF)
cv2_imshow(erosion)
cv2.waitKey(0)
```

```python
dilation = cv2.dilate(canny_edge_HPF, (5,5), iterations=3)
cv2_imshow( erosion)
cv2_imshow(dilation)
cv2_imshow(canny_edge_HPF)
cv2.waitKey(0)
```

```python
img = cv2.imread(r"/content/drive/MyDrive/Resources/garden.jpg")
gra = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gra, (5,5), 3)
canny = cv2.Canny(blur, 127, 127)
ret, thresh = cv2.threshold(blur, 127, 255, cv2.THRESH_BINARY)
# cv2.imshow('threshold', thresh)

contour, hierarchies = cv2.findContours(canny, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
print("total number of contours : ", contour.__len__())
cv2.waitKey(0)
```

```python
blank = np.zeros((img.shape), np.uint8)
cv2.drawContours(blank, contour, -1, (0, 255, 255), 2)
cv2_imshow( blank)
cv2.waitKey(0)
```

```python

# Python program to demonstrate erosion and
# dilation of images.
import cv2
import numpy as np
 
# Reading the input image
img = cv2.imread('/content/drive/MyDrive/Resources/bmw.jpg', 0)
 
# Taking a matrix of size 5 as the kernel
kernel = np.ones((5,5), np.uint8)
 
# The first parameter is the original image,
# kernel is the matrix with which image is
# convolved and third parameter is the number
# of iterations, which will determine how much
# you want to erode/dilate a given image.
img_erosion = cv2.erode(img, kernel, iterations=1)
img_dilation = cv2.dilate(img, kernel, iterations=1)
 
cv2_imshow( img)
cv2_imshow( img_erosion)
cv2_imshow(img_dilation)
```
